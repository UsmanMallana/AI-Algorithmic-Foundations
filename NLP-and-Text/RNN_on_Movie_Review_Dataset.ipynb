```python
from keras.datasets import imdb
from keras.preprocessing import sequence
import tensorflow as tf
import os
import numpy as np

VOCAB_SIZE = 88584
MAXLEN = 250
BATCH_SIZE = 64

(train_data, train_label), (test_data, test_labels) = imdb.load_data(num_words = VOCAB_SIZE)
```

```python
train_data[0]
```

```python
train_data = sequence.pad_sequences(train_data, MAXLEN)
test_data = sequence.pad_sequences(test_data, MAXLEN)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(VOCAB_SIZE, 32),
    tf.keras.layers.LSTM(32),
    tf.keras.layers.Dense(1, activation="sigmoid")
])
```

```python
model.summary()
```

```python
model.compile(optimizer="rmsprop",loss='binary_crossentropy',metrics=['acc'])
```

```python
history = model.fit(train_data, train_label, epochs=10, validation_split=0.2 )
```

```python
import keras
word_index = imdb.get_word_index()
def encode_text(text):
  tokens = keras.preprocessing.text.text_to_word_sequence(text)
  tokens = [word_index[word] if word in word_index else 0 for word in tokens]
  return sequence.pad_sequences([tokens], MAXLEN)[0]
text = "That movie was just amazing, so amazing!"
encoded = encode_text(text)
print(encoded)
```

```python
# Decoder Function
reverse_word_index = {value: key for (key, value) in word_index.items()}
def decode_integers(integers):
  PAD = 0
  text = ""
  for num in integers:
    if num != PAD:
      text += reverse_word_index[num] + " "
  return text[:-1]
print(decode_integers(encoded))
```

```python
def predict(text):
  encoded_text = encode_text(text)
  pred = np.zeros((1,250))
  pred[0] = encoded_text
  result = model.predict(pred)
  print(result[0])
```

```python
negative = "This movie was not great, it was not worth watching"
predict(negative)
```

```python
positive = "That movie was awesome, i really loved it, would watch it again, it was amazing"
predict(positive)
```
